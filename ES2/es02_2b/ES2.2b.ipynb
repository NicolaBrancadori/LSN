## Esercizio 2.2b

I parametri scelti sono identici a quelli dell' esercizio precedente:
- a=1 passo del random walk.
- $M=10^4$ e $n=100$ ripetizioni del random walk e blocchi rispettivamente.

In questo caso le direzioni possibili sono inifinite, non sono vincolato ad un reticolo cubico. Quindi devo generare in modo casuale una direzione nello spazio ovvero due angoli $\theta \in (0,\pi)$ e $\phi \in (0,2\pi)$. A questo punto ricavo le coordinate cartesiane e ne calcolo la norma.

I risultati sono riportati come in precedente, quindi 100 random walk differenti ognuno con N=1,2,...100 passi. Eseguo il fit con la funzione $f(N)=k \sqrt N$ ottendendo un k=0.92.



```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit 

def func(x,k):
	return k*np.sqrt(x)

with open("Risultati.dat") as out:
    Dati=np.loadtxt(out,usecols=(0,1),dtype=float)
    
x=np.arange(100)

k_opt=curve_fit(func,x,Dati[:,0]) # questo Ã¨ il fit. 
y_fit=func(x,k_opt[0]) #calcolo le y con i dati trovati con il fit

print("questo Ã¨ il k ottenuto con il fit: ",k_opt[0])

plt.figure(figsize=(10,10))
plt.errorbar(x,Dati[:,0],yerr=Dati[:,1],color='darkolivegreen',ecolor='blueviolet',fmt='o',markersize=5,label='Dati') 
plt.plot(x,y_fit,c='black',label="fit")
plt.xlabel('blocchi',fontsize='20')
plt.ylabel('$\sqrt{|r_N|^2}$',fontsize='20')
plt.legend(loc='upper left',fontsize='15')
plt.grid(True)
plt.show()
```

Anche in questo caso il fit Ã¨ ottimo e si vede bene il comportamneto diffusivo del random walk.
