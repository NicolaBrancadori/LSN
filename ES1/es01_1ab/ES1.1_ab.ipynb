Misuro i valori dei seguenti integrali che gi√† conosco analiticamente:

$$
\langle r \rangle = \int_0^1 r dr = 1/2 \hspace{1cm}
\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12
$$

Generando $M=10^5$ lanci e dividendoli in $N=10^2$ blocchi per ottenere un incertezza sul valore finale.
Riporto i risultati con le incertezze in funzione del numero di blocchi. All'aumentare del numero di blocchi i valori medi si avvicinano al risultato atteso e le incertezze diminuiscono.

```python
import matplotlib.pyplot as plt
import numpy as np

with open("risultati_a.dat") as out_a:
    Dati_a= np.loadtxt(out_a,usecols=(0,1),dtype=float)

with open("risultati_b.dat") as out_b:
    Dati_b= np.loadtxt(out_b,usecols=(0,1),dtype=float)
    
array_x=np.arange(100) #creo il vettore x di numeri da 1 a 100
array_x*=1000

#faccio il plot
plt.errorbar(array_x,Dati_a[:,0]-(1/12),yerr=Dati_a[:,1]) 
plt.xlabel('Lanci')
plt.ylabel('<r>-0.5')
plt.grid(True)
plt.show()

plt.errorbar(array_x,Dati_b[:,0]-(1/12),yerr=Dati_b[:,1]) 
plt.xlabel('Lanci')
plt.ylabel('(<r>-0.5)^2-1/12')
plt.grid(True)
plt.show()
```
