### Esercizio 1.3

Ho simulato l'esperimento di Buffon utilizzando come pavimento l'intervallo dell' asse reale (0,1) e per simulare le righe ho diviso tale intervallo in 100 sottointervalli e quindi le righe sono distanziate di $d=0.01$

La lunghezza dell' ago l'ho fissa a $L=0.0095$, deve essere dello stesso ordine di grandezza della distanza, ma un po' più piccola. Se fosse maggiore l'ago intersecherebbe sempre almeno una riga, se fosse troppo minore l' ago non intersecherebbe mai le righe.

Per simulare il lancio dell' ago non mi serve altro che orientare in modo casuale nello spazio la lunghezza $L$, quindi ho campionato tramite rejectin sampling l'angolo $[0,2\pi]$ nel seguente modo:
1. Genero due numeri casuali $x,y\in(-1,1)$ se $x^2+y^2<1$ accetto e 

$$
\begin{cases}
  \theta = cos^-1(\frac{x}{\sqrt{x^2+y^2}}) \hspace{0.2 cm} se \hspace{0.2 cm} y \geq 0\\
  \theta = 2\pi - cos^-1(\frac{x}{\sqrt{x^2+y^2}}) \hspace{0.2 cm} se \hspace{0.2 cm} y<0 \\
\end{cases}
$$

altrimenti se $x^2+y^2>1$ rifiuto.

Una volta che $\theta$ e $L$ posso controllare ognivolta che effettuo un lancio se questo interseca una linea e cosi stimare il valore di $\pi$ con la seguente formula:

$$ 
\pi = \frac{2L}{Pd} = \lim_{N_{lan} \to \infty}\frac{2LN_{lan}}{N_{int}d}
$$

Dove $P=frac{N_lan}{N_int}$ è al probabilità che l'ago colpisca una riga e infatti è uguale al numero di lanci diviso il numero di intersezioni.

Ho scelto di lanciare l'ago N_lan=$10^4$ volte e di dividere questi lanci in $N=10^2$ blocchi per valutare l'incertezza sul risultato. Riporto di seguito l'andamento del valore medio e della sua incertezza al variare del numero di blocchi.

```python
import matplotlib.pyplot as plt
import numpy as np

with open("risultati.dat") as out:
    Dati=np.loadtxt(out,usecols=(0,1),dtype=float)

x=np.arange(0,100)    


plt.errorbar(x,Dati[:,0],yerr=Dati[:,1]) 
plt.title('$\pi$ and $\sigma$ estimations')
plt.xlabel('blocchi')
plt.ylabel(r'$\langle \pi \rangle$')
plt.grid(True)
plt.show()
```
