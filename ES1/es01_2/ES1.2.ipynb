## Esercizio 1.2

Ho generalizzato il generatore di numeri randomici implementando due metodi: 
1. Il primo campiona una distribuzione esponenziale nel seguente modo

$$
p(x) = \lambda exp(-\lambda x), \hspace{0.5cm} x \in [0,+ \infty]
$$


2. Il secondo invece campiona una distribuzione Lorentziana:

$$
p(x) = \frac{1}{\pi} \frac{\Gamma}{(x-\mu)^2+\Gamma^2}, \hspace{0.5cm} x\in [-\infty,+\infty]
$$

Per dimostrare il TLC ho realizzato quattro istogrammi ognuno riempito con $10^4$ valori di $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (for $N=1, 2, 10, 100$) per 3 distribuzioni:
1. Gaussiana con $\mu=0$ a $\sigma=2$
2. Lorentziana con $\mu=0$ e $\Gamma=1$
3. Esponenziale con $\lambda=1$

Quello che si vede dai risultati, come il TLC enuncia, per $N \rightarrow \infty$ le medie $S_N$ tendono a distribuirsi in modo Gaussiano qualsiasi sia la distribuzione delle variabili (se sono a varianza finita e indipendenti) su cui Ã¨ fatta la media. Infatti guardando gli istogrami sottostanti, mano a mano che N aumenta le  distribuzioni esponenziale e gaussiana tendono alla loro distribuzione limite, ovvero a una Gaussiana. Mentre la Lorentziana avendo varianza infinita non rispetta le ipotesi del TCL e converge verso se stessa.

```python
import matplotlib.pyplot as plt
import numpy as np

with open("espo.dat") as outEspo:
    DatiEspo=np.loadtxt(outEspo,usecols=(0,1,2,3),dtype=float)
    

plt.hist(DatiEspo[:,0],bins=np.linspace(0,6,100))
plt.title('N=1 Dado Esponenziale')
plt.ylabel('counts')
plt.xlabel(r'$S_N$')
plt.show()

plt.hist(DatiEspo[:,1],bins=np.linspace(0,5,100))
plt.title('N=2 Dado Esponenziale')
plt.ylabel('counts')
plt.xlabel(r'$S_N$')
plt.show()

plt.hist(DatiEspo[:,2],bins=np.linspace(0,2.5,100))
plt.title('N=10 Dado Esponenziale')
plt.ylabel('counts')
plt.xlabel(r'$S_N$')
plt.show()


plt.hist(DatiEspo[:,3],bins=np.linspace(0.5,1.4,100))
plt.title('N=100 Dado Esponenziale')
plt.ylabel('counts')
plt.xlabel(r'$S_N$')


plt.show()
```

```python
with open("gauss.dat") as outGauss:
    DatiGauss=np.loadtxt(outGauss,usecols=(0,1,2,3),dtype=float)
    

plt.hist(DatiGauss[:,0],bins=np.linspace(-5,5,100))
plt.title('N=1 Dado standard')
plt.ylabel('counts')
plt.xlabel(r'$S_N$')
plt.show()

plt.hist(DatiGauss[:,1],bins=np.linspace(-5,5,100))
plt.title('N=2 Dado standard')
plt.ylabel('counts')
plt.xlabel(r'$S_N$')
plt.show()

plt.hist(DatiGauss[:,2],bins=np.linspace(-5,5,100))
plt.title('N=10 Dado standard')
plt.ylabel('counts')
plt.xlabel(r'$S_N$')
plt.show()


plt.hist(DatiGauss[:,3],bins=np.linspace(-5,5,100))
plt.title('N=100 Dado standard')
plt.ylabel('counts')
plt.xlabel(r'$S_N$')


plt.show()
```

Nel caso di una Gaussiana la distribuzione limite Ã¨ giÃ  se stessa, quello che succede Ã¨ che la deviazione standard delle medie essendo $\sigma_{N}=\sigma / \sqrt N$ diminuisce all'aumentare di $N$ e la curva infatti Ã¨ sempre piÃ¹ piccata sul valor medio $\mu=0$
