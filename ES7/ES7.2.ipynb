Fissato il numero di blocchi a 100 e il numero di passi MC all' interno di ogni blocco a 1000 completo MTV codice implementando il calcolo della funzione di distribuzione radiale $g(r)$. 

Questa funzione dà la probabilità di trovare  una coppia di atomi a distanza r rispetto alla probabilità attesa di trovarli alla stessa distanza per una distribuzione completamente casuale alla stessa densità. L' utilità di questa funzione è che da essa si può ricavare qualsiasi grandezza a due corpi come per esempio l' energia potenziale.

Per misurarla ho implementato un istogramma dividendo metà lato del cubo in 100 bins e contando il numero di particelle all' interno di ogni bins e normalizzando:
$$
g(r)=\frac{n_{part}(r)}{\rho N \Delta V(r) }
$$

dove $r \in [0,L/2]$ in modo discreto (r sono i bins), $n_{part}$ è il numero di particella all'interno di ogni bins, $\rho$ è la densità, N in numero di particelle totale e $V(r)=\frac{4 \pi}{3}[(r+dr)^3-r^3]$ potenziale di interazione di due particelle posizionate agli estremi di ogni bins.

Equilibrio il sistema con 1000 MC passi e poi eseguo la misura di $g(r)$ con 100 blocchi e 1000 passi per blocco. Riporto i valori medi per ogni blocco nel file "output.gofr.0" e i valori medi del blocco finale con le incertezze nel file "output.gave.0".

Di seguito plotto i valori finali della funzione di distribuzione radiale per i tre stati.

Dai risultati si evince che gli atomi non sono affatto distribuiti in modo casuale, ma si posizionano a distanze precise infatti la funzione di distribuzione radiale presenta dei picchi.

```python
import numpy as np
import matplotlib.pyplot as plt

with open("NVT_data/gv/output.gave.solid") as s_out:
    SG=np.loadtxt(s_out,usecols=(0,1))
with open("NVT_data/gv/output.gave.liquid") as l_out:
    LG=np.loadtxt(l_out,usecols=(0,1))
with open("NVT_data/gv/output.gave.gas") as g_out:
    GG=np.loadtxt(g_out,usecols=(0,1))

x=np.arange(SG[:,0].size)

plt.figure(figsize=(10,6))
plt.title("Fase solida")
plt.plot(x,SG[:,0])
plt.xlabel("r",fontsize=15)
plt.ylabel("g(r)",fontsize=15)
plt.show()

plt.figure(figsize=(10,6))
plt.title("Fase liquida")
plt.plot(x,LG[:,0])
plt.xlabel("r",fontsize=15)
plt.ylabel("g(r)",fontsize=15)
plt.show()

plt.figure(figsize=(10,6))
plt.title("Fase gassosa")
plt.plot(x,GG[:,0])
plt.xlabel("r",fontsize=15)
plt.ylabel("g(r)",fontsize=15)
plt.show()
```
